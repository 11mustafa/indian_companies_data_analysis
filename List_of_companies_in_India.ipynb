```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df=pd.read_csv("List_of_companies_in_India.csv")
df
```

```python
companies=df.iloc[0:100]
```

```python
companies.iloc[0:100]
```

```python
# Display basic information about the dataset
print("Dataset Information:")
print(companies.info())
```

```python
# Preview the dataset
print("First 5 Rows:")
companies.head()
```

```python
# Check for missing values
print("Checking missing values:")
print(companies.isnull().sum())
```

```python
# Fill missing values with a placeholder
companies.fillna("Unknown", inplace=True)
```

```python
# Verify missing values are handled
print("Missing values handled:")
print(companies.isnull().sum())
```

```python
# Remove duplicate records
print("Before removing duplicates:", len(companies))
companies.drop_duplicates(inplace=True)
print("After removing duplicates:", len(companies))
```

```python
# Display basic statistics about the dataset
print("Basic Statistics:")
companies.describe(include='all')
```

```python
# 1. Bar Chart: Top 10 Sectors with the Most Companies
# Question: Which sectors have the most companies?
sector_counts = companies['name'].value_counts()
sector_counts.head(15).plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title('Top 10 Sectors by Number of Companies')
plt.xlabel('name')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.show()
```

```python
if 'name' in companies.columns and 'rating' in companies.columns:
    top_rating = companies.groupby('name')['rating'].sum().sort_values(ascending=True).head(15)
    
    # Plot the chart
    plt.figure(figsize=(10, 5))
    plt.plot(top_rating.index, top_rating.values, marker='o', color='green', linestyle='-', alpha=0.7)
    plt.title('Total', fontsize=20)
    plt.xlabel('name', fontsize=15)
    plt.ylabel('rating', fontsize=15)
    plt.xticks(rotation=45, fontsize=10)
    plt.grid(alpha=0.3)
    plt.show()
else:
    print("Required columns are missing. Verify dataset structure.")
```

```python
if 'name' in companies.columns and 'reviews' in companies.columns:
    companies_reviews= companies.groupby('name')['reviews'].sum().sort_values(ascending=False).head(15)
    
    # Plot the chart
    plt.figure(figsize=(10, 5))
    plt.plot(companies_reviews.index, companies_reviews.values, marker='o', color='green', linestyle='-', alpha=0.7)
    plt.title('Total', fontsize=20)
    plt.xlabel('name', fontsize=15)
    plt.ylabel('reviews', fontsize=15)
    plt.xticks(rotation=45, fontsize=10)
    plt.grid(alpha=0.3)
    plt.show()
else:
    print("Required columns are missing. Verify dataset structure.")
```

```python
# 2. Pie Chart: Distribution of Companies by Region
# Question: How are companies distributed across different regions?
region_counts = companies['location'].value_counts().head(15)
region_counts.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribution of Companies by Region')
plt.ylabel('')
plt.show()
```
